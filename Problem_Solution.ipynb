```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r'driver-data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
features = df.values[:,1:]
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2)
```

```python
model.fit(features)
```

```python
model.labels_
```

```python
np.shape(model.labels_)
```

```python
unique,counts = np.unique(model.labels_,return_counts=True)
print(dict(zip(unique,counts)))
```

```python
data_analyze = pd.DataFrame(features,columns=['mean_dist_day','mean_over_speed_perc'])
data_analyze['cluster'] = model.labels_
sns.set_style('whitegrid')
sns.lmplot('mean_dist_day','mean_over_speed_perc',data=data_analyze,hue='cluster',palette='coolwarm',size=6,aspect=1,fit_reg=False)
```

```python
Kmeans_4 = KMeans(n_clusters=4)
Kmeans_4.fit(features)
```

```python
Kmeans_4.labels_
```

```python
unique,count=np.unique(Kmeans_4.labels_,return_counts=True)
print(dict(zip(unique,count)))
```

```python
data_analyze2 = pd.DataFrame(features,columns=['mean_dist_day','mean_over_speed_perc'])
data_analyze2['cluster'] = Kmeans_4.labels_
sns.set_style('whitegrid')
sns.lmplot('mean_dist_day','mean_over_speed_perc',data=data_analyze2,hue='cluster',palette='coolwarm',size=6,aspect=1,fit_reg=False)
```
